```python
## creat the function for finding greater number between tow numbers

def find_greater(x,y):
    if x>y:
        return(x)

    else :
        return(y)

find_greater(15,18)

## how this value but in the list :

l1=[]

def greater_num(x,y):
    if x>y:
        l1.append(x)

    else:
        l1.append(y)

greater_num(500,720)
    
l1
```

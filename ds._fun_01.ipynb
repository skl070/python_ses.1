```python
def calc(a, b, op):
    if op == '1':
        return a + b
    elif op == '2':
        return a - b
    elif op == '3':
        return a * b
    elif op == '4':
        return a / b

print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

op = input("Enter your choice: ")
a = float(input("Enter a: "))
b = float(input("Enter b: "))

if op == '4' and b == 0:
    print("Error: Division by zero")
else:
    print(calc(a, b, op))

```

```python
def add():
    a = eval(input("Enter the value of a: "))
    b = eval(input("Enter the value of b: "))
    print(a + b)

def sub():
    a = eval(input("Enter the value of a: "))
    b = eval(input("Enter the value of b: "))
    print(a - b)

def mul():
    a = eval(input("Enter the value of a: "))
    b = eval(input("Enter the value of b: "))
    print(a * b)

def div():
    a = eval(input("Enter the value of a: "))
    b = eval(input("Enter the value of b: "))
    if b == 0:
        print("Error: Division by zero")
    else:
        print(a / b)

def main():
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")

    user = input("Enter your choice: ")

    if user == '1':
        add()
    elif user == '2':
        sub()
    elif user == '3':
        mul()
    elif user == '4':
        div()
    else:
        print("Invalid choice")

main()


```

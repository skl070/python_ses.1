```python

## wap to identify the longest word and shortest word in a given sentence
sentence = "hai how are you doing today"
words = sentence.split()

longest = shortest = words[0]

for word in words:
    if len(word) > len(longest):
        longest = word
    if len(word) < len(shortest):
        shortest = word
print("Longest word:", longest)
print("Shortest word:", shortest)
```

```python
sentence = "how are you , how  going on "
words= sentence.split()

long=short=words[0]

for w in words:
    if len(w)>len(long):
        long= w
    else:
        len(w)<len(short)
        short = w

print("Lonest word:",long)
print("shortest word:",short)

```

```python
## wap take 5 random number 

# and create a dictionary with even and odd


import random

d1={}

even =[]
odd =[]


for i in range(5):
    num=random.randint(0,25)
    
    if num%2 ==0:
        even.append(num)
    else :
        odd.append(num)

print(even)
print(odd)


d1['even'] = even  # Add to dictionary
d1['odd'] = odd

print("Dictionary:", d1)

```

```python
## isomg 

s2 ='virat.kohli@blr.com,Rohit.sharma@mumbai.com,kl.rahul@lucknow.com'
s1 =s2.split(',')
first_name = []
second_name= []
company = []

d2={}
for i in s1:
    first_name.append(i[:i.find('.')])
    second_name.append(i[i.find('.')+1:i.find('@')])
    company.append(i[i.find('@')+1:i.find('.',i.find('.')+1)])


d2['first_name']=first_name
d2['second_name']=second_name
d2['company']=company

d2
```

```python
s1='virat.kohli@blr.com,Rohit.sharma@mumbai.com,kl.rahul@lucknow.com'

s2= s1.split(',')

first_n =[]
second_n=[]
last_n=[]

di={}

for i in s2:
    first_n.append (i[:i.find(".")])
    second_n.append(i[i.find(".")+1:i.find("@")])
    last_n.append(i[i.find("@")+1:i.find(".",i.find(".")+1)])


di['first_n'] =first_n
di['second_n'] =second_n
di['last_n'] =last_n

di

```
